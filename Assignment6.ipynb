```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = "https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv"
df1 = pd.read_csv(data)
```

```python
df1
```

```python
maleCount_ = 0
femaleCount_ = 0
for i in range (len(df1.loc[:,'sex'])):
    if df1.loc[i,'sex'] == 'male':
        maleCount_ += 1
    else:
        femaleCount_ += 1
```

```python
# Data to plot
labels = ['Male', 'Female']
sizes = [maleCount_, femaleCount_]
colors = ['gold', 'yellowgreen']
explode = (0.1, 0)  # explode 1st slice
 
# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=140)
 
plt.axis('equal')
plt.show()
```

```python
df = pd.DataFrame({'From_To': ['LoNDon_paris', 'MAdrid_miLAN', 'londON_StockhOlm',
'Budapest_PaRis', 'Brussels_londOn'],
'FlightNumber': [10045, np.nan, 10065, np.nan, 10085],
'RecentDelays': [[23, 47], [], [24, 43, 87], [13], [67, 32]],
'Airline': ['KLM(!)', '<Air France> (12)', '(British Airways. )', '12. Air France', '"Swiss Air"']})
```

```python
df
```

```python
for i in range(len(df['FlightNumber'])):
    if pd.isnull(df.loc[i, 'FlightNumber']):
        df.loc[i, 'FlightNumber'] = (df.loc[i - 1, 'FlightNumber'] + df.loc[i + 1, 'FlightNumber']) / 2
    
```

```python
df
```

```python
From = []
To = []
print(From)
for i in range(len(df['From_To'])):
    x = df.loc[i, 'From_To'].split('_')
    From.append(x[0])  
    To.append(x[1])
```

```python
dfTemporary = pd.DataFrame()
dfTemporary['From'] = From
dfTemporary['To'] = To
```

```python
dfTemporary
```

```python
for i in range(len(dfTemporary['From'])):
    x = dfTemporary.loc[i,'From']
    y = dfTemporary.loc[i,'To']
    dfTemporary.loc[i,'From'] = x[0].upper()+ x[1:].lower()
    dfTemporary.loc[i,'To'] = y[0].upper()+ y[1:].lower()
```

```python
dfTemporary
```

```python
df.drop(columns='From_To', inplace=True)
```

```python
df = df.join(dfTemporary)
df
```

```python
Delay_1 = []
Delay_2 = []
Delay_3 = []
for i in range(len(df['RecentDelays'])):
    val = df.loc[i, 'RecentDelays']
    if len(val) == 0:
            Delay_1.append(np.nan)
            Delay_2.append(np.nan)
            Delay_3.append(np.nan)
    elif len(val) == 1:
            Delay_1.append(val[0])
            Delay_2.append(np.nan)
            Delay_3.append(np.nan)
    elif len(val) == 2:
            Delay_1.append(val[0])
            Delay_2.append(val[1])
            Delay_3.append(np.nan)
    elif len(val) == 3:
            Delay_1.append(val[0])
            Delay_2.append(val[1])
            Delay_3.append(val[2])

        
```

```python
df.drop(columns='RecentDelays', inplace=True)
df['Delay_1'] = Delay_1
df['Delay_2'] = Delay_2
df['Delay_3'] = Delay_3
```
