```python
value = [3, 5, 7, 2, 8, 10, 11, 65, 72, 81, 99, 100, 150]
out = []
def movingAverage(value, k):
    movAvg = len(value) - k + 1
    i = 0    
    out =[]
    for x in range(0,movAvg):
        sum_ = 0
        for y in range(x, k+i):
            sum_ = sum_ + value[y] 
        out.append(sum_/k)  
        i = i+1
    return out
        
                         
out = movingAverage(value, 3)
print(out)    
    
```

```python
data = []
start_date = "2015-01-01"
end_date = "2015-12-31"

d1 = pd.date_range(start_date, end_date, freq="B")

for x in range(len(d1)):    
    data.append(str(np.random.randint(1,100))) 

df1 = pd.DataFrame(data, index = d1,columns=['MyColumn'])
```

```python
dateTimeval = pd.DatetimeIndex(df1.index.values)
total = 0

for indx in range(len(dateTimeval)):
    if dateTimeval[indx].day_name() == "Wednesday":
        total += int(df1.at[dateTimeval[indx], 'MyColumn'])
        
print("sum of s values in very wednesday in the year 2015: " , total)
```

```python
currMonth = 1
avg = 0
i = 1
averageOut = {}
for indx in range(len(dateTimeval)):    
    
    if dateTimeval[indx].month == 12 and dateTimeval[indx].day == 31:
        avg += int(df1.at[dateTimeval[indx], 'MyColumn']) / i    
        averageOut.update({dateTimeval[indx].month_name():avg})         
    elif dateTimeval[indx].month == currMonth:
        avg += int(df1.at[dateTimeval[indx], 'MyColumn']) / i        
        currMonth = dateTimeval[indx].month
        i+=1        
    elif dateTimeval[indx].month != currMonth:
        averageOut.update({dateTimeval[indx-1].month_name():avg})        
        avg = 0
        i = 1
        avg += int(df1.at[dateTimeval[indx], 'MyColumn']) / i
        currMonth = dateTimeval[indx].month
             
print(averageOut)
```
